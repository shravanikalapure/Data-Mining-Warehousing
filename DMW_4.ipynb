```python
# Name: Shravani Kalapure
# PRN: 123B1B064
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = [
    ['pen', 'pencil', 'scale'],
    ['eraser', 'pencil'],
    ['pen', 'pencil', 'scale'],
    ['eraser', 'pencil'],
    ['pen', 'pencil']
]
transactions
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
# Generate frequent itemsets
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

# Filter: Only itemsets with more than one item
frequent_itemsets_multi = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]

print(frequent_itemsets_multi)
```

```python
# Generate Association Rules
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.6)

# Show Support, Confidence, Lift
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

print("Frequent Itemsets:")
print(frequent_itemsets)

print("\n Association Rules:")
print(rules)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

transactions = [
    ['milk', 'bread', 'eggs'],
    ['milk', 'bread'],
    ['milk', 'diapers', 'beer', 'bread'],
    ['milk', 'diapers', 'bread', 'eggs'],
    ['bread', 'butter'],
    ['beer', 'chips'],
    ['bread', 'butter', 'jam'],
    ['milk', 'bread', 'butter'],
    ['milk', 'diapers', 'beer'],
    ['bread', 'diapers'],
    ['milk', 'bread', 'diapers'],
    ['milk', 'bread', 'diapers', 'beer'],
    ['diapers', 'beer'],
    ['milk', 'bread', 'butter', 'jam'],
    ['milk', 'bread', 'jam'],
    ['milk', 'bread', 'eggs', 'butter'],
    ['bread', 'jam'],
    ['milk', 'bread', 'chips'],
    ['bread', 'diapers', 'eggs'],
    ['milk', 'bread', 'diapers', 'butter'],
]
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets_multi = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
print("Frequent Itemsets :")
print(frequent_itemsets_multi)

print("\nAssociation Rules:")
print(rules)
```
